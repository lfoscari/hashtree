```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
magic = pd.read_csv("../data/magic/magic04.out")
x, y = magic['m'], magic['b']
```

```python
fig = plt.figure(figsize=(20, 20))

# Mean
ax = fig.add_subplot(1, 2, 1, projection='3d')
z = magic['avg_usage']
ax.scatter(x, y, z)

ax.set_xlabel('m')
ax.set_xbound(0, 50)

ax.set_ylabel('b')
ax.set_ybound(0, 50)

ax.set_zlabel('Usage (mean)')
ax.set_zbound(0.0, 1.0)

# Standard deviation
ax = fig.add_subplot(1, 2, 2, projection='3d')
z = magic['std_usage']
ax.scatter(x, y, z)

ax.set_xlabel('m')
ax.set_xbound(0, 50)

ax.set_ylabel('b')
ax.set_ybound(0, 50)

ax.set_zlabel('Usage (standard deviation)')

plt.show()
```

```python
fig = plt.figure()

# Mean
ax = plt.axes(projection='3d')
z = magic['avg_access']
ax.scatter(x, y, z)

ax.set_xlabel('m')
ax.set_xbound(0, 50)

ax.set_ylabel('b')
ax.set_ybound(0, 50)

ax.set_zlabel('Usage (mean)')
ax.set_zbound(0.0, 1.0)

plt.show()
```

```python
# 2D heatmap, looks a bit better, not sure if it's a good idea...

fig = plt.figure(figsize=(20, 20))

# Mean
z = magic['avg_depth']

ax = fig.add_subplot(2, 2, 1)
ax.scatter(x, y, c=z, marker=".")

ax.set_xlabel('m')
ax.set_xbound(0, 51)

ax.set_ylabel('b')
ax.set_ybound(0, 51)

# Standard deviation
z = magic['std_depth']

ax = fig.add_subplot(2, 2, 2)
ax.scatter(x, y, c=z, marker=".")

ax.set_xlabel('m')
ax.set_xbound(0, 51)

ax.set_ylabel('b')
ax.set_ybound(0, 51)

plt.show()
```

```python
def heatmap(n, heat):
    return [heat[i:i + n] for i in range(0, len(heat), n)]
    
fig = px.imshow(heatmap(max(x), z))
fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=go.Heatmap(z=z, x=x, y=y[::-1], colorscale='Viridis'))
fig.show()
```
