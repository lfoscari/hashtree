```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
magic = pd.read_csv("../data/magic/magic04.out")
x, y = magic['m'], magic['b']
```

```python
fig = plt.figure(figsize=(20, 20))

# Mean
ax = fig.add_subplot(1, 2, 1, projection='3d')
z = magic['avg_usage']
ax.scatter(x, y, z)

ax.set_xlabel('m')
ax.set_xbound(0, 50)

ax.set_ylabel('b')
ax.set_ybound(0, 50)

ax.set_zlabel('Usage (mean)')
ax.set_zbound(0.0, 1.0)

# Standard deviation
ax = fig.add_subplot(1, 2, 2, projection='3d')
z = magic['std_usage']
ax.scatter(x, y, z)

ax.set_xlabel('m')
ax.set_xbound(0, 50)

ax.set_ylabel('b')
ax.set_ybound(0, 50)

ax.set_zlabel('Usage (standard deviation)')

plt.show()
```

```python
fig = plt.figure(figsize=(20, 20))

# Mean
z = magic['avg_depth']

ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.scatter(x, y, z)

ax.set_xlabel('m')
ax.set_xbound(0, 50)

ax.set_ylabel('b')
ax.set_ybound(0, 50)

ax.set_zlabel('Depth (mean)')
ax.set_zbound(0.0, 1.0)

# Standard deviation
z = magic['std_depth']

ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.scatter(x, y, z)

ax.set_xlabel('m')
ax.set_xbound(0, 50)

ax.set_ylabel('b')
ax.set_ybound(0, 50)

ax.set_zlabel('Depth (standard deviation)')

plt.show()
```
